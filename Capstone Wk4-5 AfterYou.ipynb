```python
# function that extracts the category of the venue
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
#clean the json and structure it into a pandas dataframe
venues = results['response']['groups'][0]['items']
    
nearby_venues = json_normalize(venues) # flatten JSON

# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
nearby_venues =nearby_venues.loc[:, filtered_columns]

# filter the category for each row
nearby_venues['venue.categories'] = nearby_venues.apply(get_category_type, axis=1)

# clean columns
nearby_venues.columns = [col.split(".")[-1] for col in nearby_venues.columns]

nearby_venues.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=200):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['BranchName', 
                  'AfterYou Latitude', 
                  'AfterYou Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
##Analyze Each Branch
# one hot encoding
au_onehot = pd.get_dummies(au_venues[['Venue Category']], prefix="", prefix_sep="")

# add neighborhood column back to dataframe
au_onehot['BranchName'] = au_venues['BranchName'] 

# move neighborhood column to the first column
fixed_columns = [au_onehot.columns[-1]] + list(au_onehot.columns[:-1])
au_onehot = au_onehot[fixed_columns]

au_onehot.head()
```

```python
au_onehot.shape
```

```python
#group rows by branch and by taking the mean of the frequency of occurrence of each category
au_grouped = au_onehot.groupby('BranchName').mean().reset_index()
au_grouped
```

```python
#confirm the new size
au_grouped.shape
```

```python
#print each neighborhood along with the top 5 most common venues
num_top_venues = 5

for hood in au_grouped['BranchName']:
    print("----"+hood+"----")
    temp = au_grouped[au_grouped['BranchName'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
##put that into a pandas dataframe
#write a function to sort the venues in descending order
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
#create the new dataframe and display the top 10 venues for each neighborhood
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['BranchName']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
au_venues_sorted = pd.DataFrame(columns=columns)
au_venues_sorted['BranchName'] = au_grouped['BranchName']

for ind in np.arange(au_grouped.shape[0]):
    au_venues_sorted.iloc[ind, 1:] = return_most_common_venues(au_grouped.iloc[ind, :], num_top_venues)

au_venues_sorted.head()
```

```python
##Run k-means to cluster the branch into 3 clusters
# set number of clusters
kclusters = 4

au_grouped_clustering = au_grouped.drop('BranchName', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(au_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
##create a new dataframe that includes the cluster as well as the top 10 venues for each neighborhood
# add clustering labels
au_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

au_merged = autable

# merge toronto_grouped with toronto_data to add latitude/longitude for each neighborhood
au_merged = au_merged.join(au_venues_sorted.set_index('BranchName'), on='BranchName')
#au_merged['Cluster_Labels'] = au_merged.Cluster_Labels.astype(int)

au_merged # check the last columns!
```

```python
#visualize the resulting clusters

# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=6)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(au_merged['Latitude'], au_merged['Longitude'], au_merged['BranchName'], au_merged['Cluster Labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
